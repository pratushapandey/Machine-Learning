### Importing Libraries

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

### Reading Data Set

```python
data_set = pd.read_csv("Salary_data.csv")

#independent Variables
X = data_set.iloc[: , :-1].values

#dependent Variables
y = data_set.iloc[: , -1].values
```

### Splitting Into Test And Training Set

```python
from sklearn.model_selection import train_test_split
X_train , X_test, y_train, y_test = train_test_split(X, y , test_size = 0.3, random_state = 42) #30% is 0.3
```

### Training on Trainning Set Using Simple Linear Regression

```python
from sklearn.linear_model import LinearRegression
regressor = LinearRegression() #regressor is the instance of the particular function
regressor.fit(X_train, y_train)
```

### Prediction On Test Set

```python
y_pred = regressor.predict(X_test)                                                                                 
```

### Visualizing The Training Set

```python
plt.scatter(X_train, y_train , color = 'red')
plt.plot(X_train, regressor.predict(X_train),color = 'blue')
plt.title("CTC VS Experience")
plt.xlabel("Experience In Years")
plt.ylabel("CTC")
```

### Visualizing The Test Set

```python
plt.scatter(X_test, y_test , color = 'red')
plt.plot(X_test, y_pred,color = 'blue')
plt.title("CTC VS Experience")
plt.xlabel("Experience In Years")
plt.ylabel("CTC")
```
